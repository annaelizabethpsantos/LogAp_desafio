```
#ler o arquivo CSV e plotar

import matplotlib.pyplot as plt
import pandas as pd
import csv

csvfile = open('/content/Abr-2017-curva-potencia-windbox.csv', 'r')
#spamreader = csv.reader(csvfile, delimiter=';')

x = []
y = []

for line in csvfile:
  line = line.strip()
  X, Y = line.split(';')
  x.append(X)
  y.append(Y)

plt.scatter(x, y, color='blue')
plt.xlabel("Velocidade do Vento")
plt.ylabel("PotÃªncia")
plt.show()
```
